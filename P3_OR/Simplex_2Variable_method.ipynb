**Two Variable Simplex Method**

Max z= 3x1 + 2x2

Subject to 

x1 + x2 <=4

x1 - x2 <=2

x1,x2 >=0

```
from scipy.optimize import linprog
```

```
obj = [-3 ,-2]
```

```
lhs_ineq = [[1, 1],
            [1, -1]]
```

```
rhs_ineq = [4,
            2]
```

```
bnd = [(0, float("inf")),
       (0, float("inf"))]
```

```
opt = linprog(c=obj, A_ub=lhs_ineq , b_ub=rhs_ineq,
              bounds=bnd,method="revised simplex")
```
