
21_SHIVANI PAL

**Resource Allocation Problem**

Max z = 20x1 + 12x2 + 40x3 +  25x4

Subject to:

x1 + x2 + x3 +x4 <=50 

3x1 + 2x2 + x3  <=100

x2 + 2x3 <=90

x1 , x2 ,x3 , x4 >=0

```
from scipy.optimize import linprog
```

```
obj = [-20 , -12 , -40 , -25]
```

```
lhs_ineq = [[1, 1, 1, 1],
            [3, 2, 1, 0],
            [0, 1, 2 ,3]]
```

```
rhs_ineq = [50,
            100,
            90]
```

```
opt = linprog(c=obj, A_ub=lhs_ineq, b_ub=rhs_ineq,
              method="revised simplex")
```
