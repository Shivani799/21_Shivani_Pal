**Three Variable Simplex Method**

Min z= x1-3x2 + 2x3

Subject to

3x1 - x2+3x3 <=7

-2x1 + 4x2 <=12

-4x1 +3x2+8x3<=10

x1,x2, x3 >=0

```
from scipy.optimize import linprog
```

```
obj = [-1 ,-3, 2]
```

```
lhs_ineq = [[3, -1, 3],
            [-2, 4, 0],
            [-4, 3, 8]]
```

```
rhs_ineq = [7,
            12,
            10]
```

```
bnd = [(0, float("inf")),
       (0, float("inf")),
       (0, float("inf"))]
```

```
opt = linprog(c=obj, A_ub=lhs_ineq , b_ub=rhs_ineq,
              bounds=bnd,method="revised simplex")
```
