**INFEASIBILITY IN SIMPLEX METHOD**

Max z = 200x - 300y 

subject to 

2x+3y>=1200

x+y<=400

2x+3/2y>=900

x,y>=0

```
from scipy.optimize import linprog
```

```
obj = [-200, 300]
```

```
lhs_ineq = [[-2, -3],
            [1, 1],
            [-2, -1.5]]
```

```
rhs_ineq = [-1200, 400, -900]
```

```
bnd = [(0, float("inf")),
       0, float("inf")]
```

```
opt = linprog(c=obj , A_ub=lhs_ineq, B_ub=rhs_ineq,
              bounds=bnd, method="revised simplex")
```
