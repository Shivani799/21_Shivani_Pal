21_SHIVANI PAL

**Big M Simplex Method**

Min z = 4x1 + x2

Subject to:

3x1 + 4x2 >=20

x1 + 5x2 >=15

x1,x2>=0

```
from scipy.optimize import linprog
```

```
obj = [4, 1]
```

```
lhs_ineq = [[-3, -4],
            [-1, -5]]
```

```
rhs_ineq = [-20, -15]
```

```
bnd = [(0, float("inf")),
       (0, float("inf"))]
```

```
opt = linprog(c=obj, A_ub=lhs_ineq, b_ub=rhs_ineq,
              bounds=bnd,method="interior-point")
```
